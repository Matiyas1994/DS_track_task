```python
import io
df = pd.read_csv(io.BytesIO(uploaded['imdb_top_1000.csv']))
```

```python
column_names = df.columns.tolist()
print("Column names:", column_names)
```

```python
average_rating = df['IMDB_Rating'].mean()
print("Average rating:", average_rating)
```

```python
unique_genres = df['Genre'].nunique()
print("Number of unique genres:", unique_genres)
```

```python
results_df = pd.DataFrame({
    'Column Names': [column_names],
    'Average Rating': [average_rating],
    'Number Of Unique Genres': [unique_genres]
})
results_df.to_csv('results.csv', index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
genre_counts = df['Genre'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(genre_counts.index, genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Genre Distribution of Movies')
plt.xticks(rotation=90)
plt.show()
```

```python
year_counts = df['Released_Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.scatter(year_counts.index, year_counts.values)
plt.xlabel('Released_Year')
plt.ylabel('Number of Movies')
plt.title('Number of Movies vs. Year')
plt.xticks(rotation=90)
plt.show()
```
